```python
import numpy as np
import matplotlib.pyplot as plt

x = ['intersection', 'lane_change', 'normal_curved_s_turn', 'stopped_traffic_signal', 'turn']
# y = [395, 351, 237, 326,118]

data = [[202, 141, 132, 172, 65],
[16, 46, 13, 24, 13],
[176, 164, 85, 130, 40]]
X = np.arange(5)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.bar(X + 0.00, data[0], color = 'b', width = 0.25)
ax.bar(X + 0.25, data[1], color = 'g', width = 0.25)
ax.bar(X + 0.50, data[2], color = 'r', width = 0.25)
ax.set_ylabel('Video Count')
ax.set_title('Number of Occurences for each class')
ax.set_xticks(X)
ax.set_xticklabels(x, rotation=45)
ax.legend(labels=['Train', 'Val', 'Test'])

plt.show()
```
